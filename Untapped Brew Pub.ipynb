```python
#import dependencies 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#store in data file
untapped_brew_pub = "Resources/untapped_brew_pub.csv"
```

```python
untapped_brew_pub_df = pd.read_csv(untapped_brew_pub)
```

```python
untapped_brew_pub_df.head()
```

```python
#top 10 brewery by ratings
best_brew = untapped_brew_pub_df.sort_values("brewery_ratings", ascending=False)
best_brew = best_brew.iloc[:10]
best_brew
```

```python
#worst brewery by ratings
worst_brew = untapped_brew_pub_df.sort_values("brewery_ratings")
worst_brew = worst_brew.iloc[:10]
worst_brew
```

```python
y_axis = np.arange(len(best_brew))
tick_locations = [value for value in y_axis]

fig0, ax0 = plt.subplots()
plt.figure(figsize=(20,10))
ax0.barh(y_axis, best_brew["brewery_ratings"], color='tab:orange',
        alpha=0.8, align="center")
ax0.set_yticks(tick_locations, best_brew["brewery_location"], rotation="horizontal")
ax0.set_title("Best Breweries in The World")
ax0.set_ylabel("Location")
ax0.set_xlabel("Brewery Ratings")
ax0.invert_yaxis()
fig0.tight_layout()
plt.show()
```
